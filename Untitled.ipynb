```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data
```

```python
data.columns
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['KK'],label = 'KK',linewidth="3.0")
plt.plot(data['SRep_Rand_No_Part'],label = 'Rep_Rand_No_Part',linewidth="3.0")
plt.plot(data['Hill_Climb_No_Part'], label = 'Hill_Climb_No_Part',linewidth="3.0")
plt.plot(data['Annealing_No_Part'],label='Annealing_No_Part',linewidth="3.0")
plt.legend(loc="upper left")
plt.title("Standard Residual")
plt.show()
plt.savefig("try1.png")
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['Rep_Rand_Part'],label = 'Rep_Rand_Part',linewidth="3.0")
plt.plot(data['Hill_Climb_Part'], label = 'Hill_Climb_Part',linewidth="3.0")
plt.plot(data['Annealing_Part'],label='Annealing_Part',linewidth="3.0")
plt.legend(loc="upper left")
plt.title("Residual with pre-partitioning")
plt.show()
#plt.savefig("try1.png")
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['SRep_Rand_No_Part'],label = 'Rep_Rand_No_Part',linewidth="3.0")
plt.plot(data['Rep_Rand_Part'], label = 'Rep_Rand_Part',linewidth="3.0")
plt.legend(loc="upper left")
plt.title("Repeated Random")
plt.show()
#plt.savefig("try1.png")
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['Hill_Climb_No_Part'],label = 'Hill_Climb_No_Part',linewidth="3.0")
plt.plot(data['Hill_Climb_Part'], label = 'Hill_Climb_Part',linewidth="3.0")
plt.legend(loc="upper left")
plt.title("Hill Climbing")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['Annealing_No_Part'],label = 'Annealing_No_Part',linewidth="3.0")
plt.plot(data['Annealing_Part'], label = 'Annealing_Part',linewidth="3.0")
plt.legend(loc="upper left")
plt.title("Annealing")
plt.show()
```
